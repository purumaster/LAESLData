```python
import pandas as pd
import sqlite3
import os
os.getcwd()

```

```python
df1 = pd.read_csv("LAESLData.csv")
df1.head(20)

# create the sqlite DB
db = sqlite3.connect("LAESLData.db")

# convert the pandas dataframe to a sqlite table
df1.to_sql("LAESL", db, if_exists="replace")

# view dataframe
# df1
query = ''' 
SELECT * 
FROM LAESL'''
pd.read_sql_query(query,db)
```

```python
# Fix Morrocoã€€to Morocco in Country.
query = '''
SELECT REPLACE(Country, "Morroco", "Morocco")
FROM LAESL;

'''

pd.read_sql_query(query,db)
```

```python
# Alphabetical order of Countries.
query = '''
SELECT DISTINCT Country 
FROM LAESL 
ORDER BY Country ASC;

'''

pd.read_sql_query(query,db)
```

```python
# Counting for how many people from each countries.
query = '''
SELECT Country, COUNT(*) AS how_many_people
FROM LAESL 
GROUP BY Country;

'''

pd.read_sql_query(query,db)
```

```python
# Fix "chinese"to "Chinese" in Language.
query = '''
SELECT REPLACE(Language, "chinese", "Chinese")
FROM LAESL;

'''

pd.read_sql_query(query,db)
```

```python
# Alphabetical order of Languges.
query = '''
SELECT DISTINCT Language 
FROM LAESL 
ORDER BY Language ASC;

'''

pd.read_sql_query(query,db)
```

```python
# Counting for how many Languages.
query = '''
SELECT Language, COUNT(*) AS how_many_Language
FROM LAESL 
GROUP BY Language;

'''

pd.read_sql_query(query,db)
```

```python
# Counting for genders.
query = '''
SELECT COUNT(*)
FROM LAESL
GROUP BY Gender = "M","F";

'''

pd.read_sql_query(query,db)
```
